```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt

# question 1
vec = np.zeros((15), int)
print(vec,'displaying vec')
# updating each item with random number with in 1 to 20
i=0
for item in vec:
    random_value = random.randrange(1,20)
    vec[i] = random_value
    i = i+1

print('after modifying : \n', vec)
output = np.reshape(vec,(3,5))
print('after reshaping: \n', output)
# finding max from each row
max = np.amax(output,axis=1)
print('max values \n', max)
output[np.arange(len(output)), np.argmax(output, axis = 1)] = 0
print("After replacing max with 0: \n", output)

# question 2
df = pd.read_csv('data.csv')
print("After reading data: ",df.head())
# statistical info
stat_info = df.describe()
print('Stat info: ',stat_info)
# checking null values
null_values = df.isnull().sum()
print("Null values in data: \n", null_values)
# replacing null with mean
df["Calories"].fillna(df["Calories"].mean(), inplace=True)
print(df["Calories"].head(122))
# aggregating Duration and pulse columns
result = df.groupby('Duration').agg({'Pulse': ['min', 'max', 'count', 'mean']})
print("After aggregation: \n",result)
#  calories values between 500 and 1000
calories_btw_500_1000 = df[(df["Calories"] > 500) & (df["Calories"] < 1000)]
print("Calories between 500 and 1000: \n",calories_btw_500_1000)
# calories values > 500 and pulse < 100
calories_500more_pulse_100less = df[(df["Calories"] > 500) & (df["Pulse"] < 100)]
print("Calories more than 500 and pulse less than 100: \n",calories_500more_pulse_100less)
# df_modified without Maxpulse column
df_modified=pd.DataFrame(df,columns=["Duration","Pulse","Calories"])
print("Df_modified: \n",df_modified.head())
# Delete the “Maxpulse” column from the main df dataframe
df.drop("Maxpulse",axis=1,inplace=True)
print("After deleting max pulse from original: \n", df.head())
# converting calories to int type
df["Calories"] = df["Calories"].astype(int)
print("After converting to int: \n", df.head())
# scatter plot for Duration and calories
df.plot.scatter(x="Duration", y="Calories", c='DarkBlue')
```
